```python
import random
import os
from IPython.display import clear_output
import time

def draw_map(player_pos, core_pos, firewalls, size=6):
    # Clear the previous screen for a clean, updating animation effect
    clear_output(wait=True)
    print("🌐 CYBER-GRID INTRUSION SIMULATOR 🌐")
    print("Commands: W (up), S (down), A (left), D (right)\n")

    for r in range(size):
        row_str = ""
        for c in range(size):
            if [r, c] == player_pos:
                row_str += " H "  # Hacker
            elif [r, c] == core_pos:
                row_str += " C "  # Core
            elif [r, c] in firewalls:
                row_str += " X "  # Firewall
            else:
                row_str += " . "  # Empty Space
        print(row_str)
    print("-" * 25)

def play_hacker():
    size = 6
    player = [0, 0]
    core = [5, 5]
    # Place random firewall obstacles
    firewalls = [[1, 2], [2, 4], [3, 1], [4, 3], [5, 2]]

    while True:
        draw_map(player, core, firewalls, size)

        if player == core:
            print("🔓 ACCESS GRANTED. Data Core downloaded! YOU WIN! 🏆")
            break
        elif player in firewalls:
            print("🚨 SYSTEM COMPROMISED. Firewall isolated your signature. GAME OVER! ☠️")
            break

        move = input("Enter direction (W/A/S/D): ").upper().strip()

        # Calculate new position based on grid logic
        if move == 'W' and player[0] > 0: player[0] -= 1
        elif move == 'S' and player[0] < size - 1: player[0] += 1
        elif move == 'A' and player[1] > 0: player[1] -= 1
        elif move == 'D' and player[1] < size - 1: player[1] += 1

play_hacker()
```
