```python
import random
import time

def start_adventure():
    print("🏰 QUEST FOR THE ANCIENT VAULT 🏰")
    print("=" * 40)
    print("You have finally reached the inner chamber of the Sunken Citadel.")
    print("In front of you stands a massive golden door with a glowing grid of letters.")
    time.sleep(1.5)

    print("\nA magical voice echoes: 'To pass, you must unscramble the password!'")
    print("The letters shift and lock into place...")
    time.sleep(1)

    # Puzzle Setup
    word_bank = {
        "treasure": "Something valuable, like gold or gems.",
        "keyhole": "The place where you insert a key.",
        "chamber": "A large room, typically underground.",
        "ancient": "Belonging to the very distant past."
    }

    secret_word = random.choice(list(word_bank.keys()))
    scrambled = "".join(random.sample(secret_word, len(secret_word)))

    print(f"\n🔮 SCRAMBLED WORD: {scrambled.upper()}")
    print(f"🔮 HINT: {word_bank[secret_word]}")
    print("-" * 40)

    guess = input("Enter your guess to unlock the door: ").lower().strip()

    if guess == secret_word:
        win_ending()
    else:
        print("\n❌ The grid flashes RED! An incorrect guess!")
        print("The floor beneath you begins to open up into a dark pit!")
        time.sleep(1.5)
        danger_choice()

def danger_choice():
    print("\n⚠️ QUICK! You only have a split second to react!")
    choice = input("Do you JUMP across the gap, or GRAB the edge of the door? (jump/grab): ").lower().strip()

    if choice == "jump":
        print("\n🏃 You dive across the gap, landing face-first on the solid stone floor just as the pit opens completely.")
        print("You look back—the door clicked open anyway out of respect for your reflexes! YOU WIN! 🎉")
    elif choice == "grab":
        print("\n🧗 You catch the edge of the door! But it's too slippery to hold onto...")
        print("You slide down into the darkness. GAME OVER! 👻")
    else:
        print("\n🔄 You froze up trying to decide! You fell into the pit. GAME OVER! 🕳️")

def win_ending():
    print("\n🟢 The grid flashes GREEN! The heavy golden door slowly rumbles open.")
    print("Inside, piles of ancient artifacts shine brightly. You successfully completed the quest! YOU WIN! 🏆")

# Start the game
start_adventure()
```
