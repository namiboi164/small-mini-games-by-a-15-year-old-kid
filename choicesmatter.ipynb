```python
import time

def intro():
    print("🌲 You wake up at the edge of a dark, whispering forest.")
    print("In front of you, the path splits into two directions.")
    time.sleep(1)

    choice = input("\nDo you go LEFT into the misty woods, or RIGHT toward the glowing cave? (left/right): ").lower().strip()

    if choice == "left":
        misty_woods()
    elif choice == "right":
        glowing_cave()
    else:
        print("Invalid choice! A wild wind blows you back to the start.")
        intro()

def misty_woods():
    print("\n🌲 You walk into the mist. It gets colder.")
    print("Suddenly, a friendly talking owl appears on a branch!")
    time.sleep(1)

    choice = input("Do you TALK to the owl, or IGNORE it and walk past? (talk/ignore): ").lower().strip()

    if choice == "talk":
        print("\n🦉 The owl tells you a secret shortcut! You safely find a chest full of gold. YOU WIN! 🎉")
    else:
        print("\n👟 You get lost in the fog and wander around in circles forever. GAME OVER. 👻")

def glowing_cave():
    print("\n🪨 You enter the cave. The walls are covered in glowing blue crystals.")
    print("Deep inside, you see a sleeping dragon guarding a glittering sword!")
    time.sleep(1)

    choice = input("Do you SNEAK past the dragon, or RUN away? (sneak/run): ").lower().strip()

    if choice == "sneak":
        print("\n🤫 You successfully grab the sword without waking the beast! You are a hero! YOU WIN! ⚔️")
    else:
        print("\n🏃 You trip over a rock while running away. The dragon wakes up... GAME OVER. 🔥")

# Start the game
intro()
```
