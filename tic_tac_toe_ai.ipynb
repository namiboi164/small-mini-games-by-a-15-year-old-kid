# 🎮 Unbeatable Tic-Tac-Toe AI

A Tic-Tac-Toe game where the AI opponent uses the **minimax algorithm** to play perfectly.
It literally cannot lose — the best you can ever do is force a draw.

**Built with:** Python (no libraries needed beyond the standard `math` module)

**Concepts used:** recursion, game trees, the minimax algorithm, basic terminal I/O

---
▶️ Run the cell below, then follow the prompts in the output box to play.


```python
import math

def print_board(board):
    symbols = {0: " ", 1: "\u274c", -1: "\u2b55"}
    print()
    for i in range(3):
        row = [symbols[board[i*3+j]] for j in range(3)]
        print(f" {row[0]} | {row[1]} | {row[2]} ")
        if i < 2:
            print("---+---+---")
    print()

def check_winner(board):
    lines = [(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    for a, b, c in lines:
        if board[a] != 0 and board[a] == board[b] == board[c]:
            return board[a]
    return 0 if 0 not in board else None

def minimax(board, depth, maximizing):
    winner = check_winner(board)
    if winner is not None:
        return {1: 10 - depth, -1: depth - 10, 0: 0}[winner]
    if maximizing:
        best = -math.inf
        for i in range(9):
            if board[i] == 0:
                board[i] = 1
                best = max(best, minimax(board, depth + 1, False))
                board[i] = 0
        return best
    else:
        best = math.inf
        for i in range(9):
            if board[i] == 0:
                board[i] = -1
                best = min(best, minimax(board, depth + 1, True))
                board[i] = 0
        return best

def best_ai_move(board):
    best_score, move = -math.inf, None
    for i in range(9):
        if board[i] == 0:
            board[i] = 1
            score = minimax(board, 0, False)
            board[i] = 0
            if score > best_score:
                best_score, move = score, i
    return move

def get_human_move(board):
    while True:
        try:
            pos = int(input("Your move (1-9): ")) - 1
            if 0 <= pos <= 8 and board[pos] == 0:
                return pos
            print("That spot is taken or out of range. Try again.")
        except ValueError:
            print("Please enter a number 1-9.")

def play_round(human_first):
    board = [0] * 9
    turn = -1 if human_first else 1
    print_board(board)
    while True:
        if turn == -1:
            board[get_human_move(board)] = -1
        else:
            print("\U0001f916 AI is thinking...")
            board[best_ai_move(board)] = 1
        print_board(board)
        result = check_winner(board)
        if result is not None:
            return result
        turn *= -1

def main():
    wins = losses = draws = 0
    print("\U0001f3ae Welcome to Unbeatable Tic-Tac-Toe! You are \u2b55, AI is \u274c.")
    while True:
        first = input("Go first? (y/n): ").strip().lower() == "y"
        result = play_round(first)
        if result == -1:
            print("\U0001f389 You won?! That should be impossible against perfect play \u2014 screenshot it!")
            wins += 1
        elif result == 1:
            print("\U0001f916 AI wins. Better luck next time!")
            losses += 1
        else:
            print("\U0001f91d Draw \u2014 that is the best anyone can do against perfect play!")
            draws += 1
        print(f"Score -> You: {wins} | AI: {losses} | Draws: {draws}")
        if input("Play again? (y/n): ").strip().lower() != "y":
            print("Thanks for playing! \U0001f44b")
            break

main()
```

## 🧠 How the AI works

Minimax explores every possible sequence of future moves and assumes both players play
optimally: the AI (maximizing player) tries to get the best possible score, while it assumes
you (minimizing player) will also always pick your best possible move. Because Tic-Tac-Toe is
small enough (at most 9! game states), the AI can look all the way to the end of the game
every single turn, so it never makes a mistake.

This is the same core idea used in game-playing engines for chess and checkers, just on a
much smaller board.
